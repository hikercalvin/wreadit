```python
from dateutil import parser
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
import pandas as pd
import re 
import numpy as np
```

```python
all_data = pd.read_csv("Pagevalue_20210101-20200810.csv")
```

```python
pv_data = pd.read_csv('系列PV綜整.csv')
```

```python
all_data_2021 = pd.read_csv('process_data_pv.csv')
```

```python
all_data_2021
```

```python
all_data_2021.info()
```

```python
all_data_2021["時間戳記"] = pd.to_datetime(all_data_2021["時間戳記"])
```

```python
all_data_2021["時間戳記_月份"] = all_data_2021["時間戳記"].dt.to_period("M")
all_data_2021
```

```python
all_data_2021["時間戳記_年份"] = all_data_2021["時間戳記"].dt.year
```

```python
series_data_month = all_data_2021.groupby(["頻道", "時間戳記_年份", "時間戳記_月份"], as_index=False)["瀏覽量"].sum()
series_data_month
```

```python
# 我們要將「時間戳記_月份」轉成字串
series_data_month['時間戳記_月份'] = series_data_month['時間戳記_月份'].astype(str) 
```

```python
series_data_month_tmp = series_data_month[series_data_month['頻道'] =='美食旅遊']
```

```python
# 再次繪製折線圖：
fig = px.line(series_data_month_tmp, 
              x="時間戳記_月份", 
              y="瀏覽量", 
            #   color = '訂單時間_年份',
              title = 'Top1_高PV系列_「美食旅遊」_歷年PV分析圖')

# 如何加入背景元素（layout）
fig.update_layout(
                title={
                    'text': 'Top1_高PV系列_「美食旅遊」_歷年PV分析圖',
                    'y':0.9,
                    'x':0.5,
                    'xanchor': 'center',
                    'yanchor': 'top'},
                xaxis_title='月份',
                yaxis_title='瀏覽量')
plot(fig, filename = 'Top1_高PV系列_「美食旅遊」_歷年PV分析圖.html')
```

```python
rank = 1
for series in pv_data["頻道"]:
        series_data_month_tmp = series_data_month[series_data_month['頻道'] ==series]

        fig = px.line(series_data_month_tmp,
            x="時間戳記_月份", 
            y="瀏覽量", 
            #   color = '訂單時間_年份',
            title = 'Top' + str(rank)+"_高PV系列_" + series +"_歷年PV分析圖")
    
        fig.update_layout(
                title={
                    'text': "Top" + str(rank)+"_高PV系列_" + series +"_歷年PV分析圖",
                    'y':0.9,
                    'x':0.5,
                    'xanchor': 'center',
                    'yanchor': 'top'},
                xaxis_title='月份',
                yaxis_title='瀏覽量')
        plot(fig, filename = "Top" + str(rank)+"_高PV系列_" + series +"_歷年PV分析圖.html")
        rank = rank + 1
```
