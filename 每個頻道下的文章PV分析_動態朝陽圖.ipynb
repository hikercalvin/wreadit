```python
import pandas as pd
import numpy as np
from datetime import timedelta, datetime
import os
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
pv_data = pd.read_csv('系列PV綜整.csv')
```

```python
all_data_2021 = pd.read_csv('process_data_pv.csv')
```

```python
all_data_2021
```

```python
all_data_2021["時間戳記"] = pd.to_datetime(all_data_2021["時間戳記"])
```

```python
all_data_2021
```

```python
pv_data
```

```python
pv_data['瀏覽量佔比'] = pv_data['瀏覽量'] / pv_data['瀏覽量'].sum()
pv_data["累計瀏覽量佔比"] = pv_data["瀏覽量佔比"].cumsum()
pv_data
```

```python
pv_data['累計頻道次數'] = range(1,len(pv_data)+1)
pv_data['累計頻道佔比'] = pv_data['累計頻道次數'] / len(pv_data)
pv_data
```

```python
# 篩選貢獻80%PV的頻道
pv_percent = 0.8
# 問題：抓取小於且等於「累計瀏覽量佔比」的頻道
pv_data_selected = pv_data[pv_data["累計瀏覽量佔比"]<=pv_percent]
```

```python
pv_data_selected
```

```python
all_data_top = []
for series in pv_data_selected["頻道"]:
    all_data_temp = all_data_2021[all_data_2021["頻道"]==series]
    all_data_top.append(all_data_temp)
```

```python
all_data_top = pd.concat(all_data_top, axis= 0)
all_data_top
```

```python
# 繪製朝陽圖
fig = px.sunburst(all_data_top, 
                  path=['頻道', '文章標題'], values='瀏覽量', color='瀏覽量')
plot(fig, filename='高PV頻道與文章標題朝陽圖.html')
```
