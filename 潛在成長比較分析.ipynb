```python
# series_data_month：每個頻道在每一個月的瀏覽量
# series:放哪一個頻道
# rank: topN，前幾個高PV頻道的排名
# compare_year1:初始年份，在此為2020
# compare_year2:最新的比較年份，在此2021

def potiential_growth(series_data_month, series, rank, compare_year1, compare_year2):
    series_data_month_growth = series_data_month[series_data_month['頻道'] == series]
    series_data_month_growth = series_data_month_growth[series_data_month_growth['時間戳記_年份'] >= int(compare_year1.strip('年'))]
    test = series_data_month_growth[series_data_month_growth["時間戳記_年份"]==int(compare_year1.strip('年'))]
    if len(test) ==0:
        print('沒有初始年份可以比較')
        test = '跳過'
        series_data_month_growth = '沒有'
        return series_data_month_growth, test
    else:
        print('有初始年份可以比較!')
        test = '有資料'
    series_data_month_growth['時間戳記_月份'] = series_data_month_growth['時間戳記_月份'].astype(str)
    series_data_month_growth['月份'] = series_data_month_growth['時間戳記_月份'].str.split('-',expand=True)[1]  #講切割出來的欄位延伸出來兩攔[0,1],然後選擇[1]
    series_data_month_growth = series_data_month_growth.pivot(index=['月份','頻道'], columns='時間戳記_年份', values='瀏覽量')
    series_data_month_growth = series_data_month_growth.reset_index()
    series_data_month_growth = series_data_month_growth.rename(columns={int(compare_year1.strip('年')):compare_year1, int(compare_year2.strip('年')):compare_year2 })
    # 四捨五入小數點
    series_data_month_growth = round(series_data_month_growth)
    # 目前只有七八月有資料
    series_data_month_growth = series_data_month_growth.iloc[[6,7],:]

    # 計算每個月的年增長
    series_data_month_growth['年增成長'] = round( (series_data_month_growth[compare_year2] - series_data_month_growth[compare_year1] )/ series_data_month_growth[compare_year1], 2)
    # 計算「總年增成長率」：了解今年比較去年來說，增長了多少
    series_data_month_growth['總年增成長率'] = (series_data_month_growth[compare_year2].sum() - series_data_month_growth[compare_year1].sum())/ series_data_month_growth[compare_year1].sum()
    series_data_month_growth.to_csv( 'Top' +  str(rank) +'_高PV頻道_「' + series + '」_「潛在」成長比較分析表.csv', encoding='utf-8-sig')
     # 繪製「潛在」成長比較分析圖



    # rank = 1 
    # series = '系列3'
    # 再次繪製折線圖：
    fig = px.line(series_data_month_growth, 
                x="月份", 
                y="年增成長", 
                #   color = '訂單時間_年份',
                title ='Top' +  str(rank) +'_高獲利系列_「' + series + '」_「潛在」成長比較分析圖')

    # 如何加入背景元素（layout）
    fig.update_layout(
                    title={
                        'text':  'Top' +  str(rank) +'_高獲利系列_「' + series + '」_「潛在」成長比較分析圖',
                        'y':0.9,
                        'x':0.5,
                        'xanchor': 'center',
                        'yanchor': 'top'},
                    xaxis_title='月份',
                    yaxis_title='年增成長')
    fig.update_traces(mode="lines+markers")
    plot(fig, filename = 'Top' +  str(rank) +'_高獲利系列_「' + series + '」_「潛在」成長比較分析圖.html')

    # rank = rank +1
    return series_data_month_growth, test

```

```python
series_data_month_growth, test = potiential_growth(series_data_month=series_data_month, series = '美食旅遊', rank = 1, compare_year1 = '2020年', compare_year2='2021年')
```

```python
# 存每一系列的「總體「潛在」成長比較分析」
series_data_month_growth_all_single = []

# 存每一系列的「全部系列「潛在」成長比較分析表」
series_data_month_growth_all = []

rank = 1
for series in for_bar_data['頻道']:
    print(series)
    series_data_month_growth, test = potiential_growth(series_data_month=series_data_month, series = series, rank = rank, compare_year1 = '2020年', compare_year2='2021年')
    rank = rank + 1 

    if test == '有資料':
        # 製作總體「潛在」成長比較分析表
        series_data_month_growth_single = series_data_month_growth[['頻道', '總年增成長率']]


        series_data_month_growth_single = series_data_month_growth_single.iloc[0:1, :]
        
        # 存每一系列的「總體「潛在」成長比較分析」
        series_data_month_growth_all_single.append(series_data_month_growth_single)
        
        # 存每一系列的「全部系列「潛在」成長比較分析表」
        series_data_month_growth_all.append(series_data_month_growth)


    
    elif test == '跳過':
        print(series, '跳過')
```
