```python
# series_data_monthï¼šæ¯å€‹é »é“åœ¨æ¯ä¸€å€‹æœˆçš„ç€è¦½é‡
# series:æ”¾å“ªä¸€å€‹é »é“
# rank: topNï¼Œå‰å¹¾å€‹é«˜PVé »é“çš„æ’å
# compare_year1:åˆå§‹å¹´ä»½ï¼Œåœ¨æ­¤ç‚º2020
# compare_year2:æœ€æ–°çš„æ¯”è¼ƒå¹´ä»½ï¼Œåœ¨æ­¤2021

def potiential_growth(series_data_month, series, rank, compare_year1, compare_year2):
    series_data_month_growth = series_data_month[series_data_month['é »é“'] == series]
    series_data_month_growth = series_data_month_growth[series_data_month_growth['æ™‚é–“æˆ³è¨˜_å¹´ä»½'] >= int(compare_year1.strip('å¹´'))]
    test = series_data_month_growth[series_data_month_growth["æ™‚é–“æˆ³è¨˜_å¹´ä»½"]==int(compare_year1.strip('å¹´'))]
    if len(test) ==0:
        print('æ²’æœ‰åˆå§‹å¹´ä»½å¯ä»¥æ¯”è¼ƒ')
        test = 'è·³é'
        series_data_month_growth = 'æ²’æœ‰'
        return series_data_month_growth, test
    else:
        print('æœ‰åˆå§‹å¹´ä»½å¯ä»¥æ¯”è¼ƒ!')
        test = 'æœ‰è³‡æ–™'
    series_data_month_growth['æ™‚é–“æˆ³è¨˜_æœˆä»½'] = series_data_month_growth['æ™‚é–“æˆ³è¨˜_æœˆä»½'].astype(str)
    series_data_month_growth['æœˆä»½'] = series_data_month_growth['æ™‚é–“æˆ³è¨˜_æœˆä»½'].str.split('-',expand=True)[1]  #è¬›åˆ‡å‰²å‡ºä¾†çš„æ¬„ä½å»¶ä¼¸å‡ºä¾†å…©æ””[0,1],ç„¶å¾Œé¸æ“‡[1]
    series_data_month_growth = series_data_month_growth.pivot(index=['æœˆä»½','é »é“'], columns='æ™‚é–“æˆ³è¨˜_å¹´ä»½', values='ç€è¦½é‡')
    series_data_month_growth = series_data_month_growth.reset_index()
    series_data_month_growth = series_data_month_growth.rename(columns={int(compare_year1.strip('å¹´')):compare_year1, int(compare_year2.strip('å¹´')):compare_year2 })
    # å››æ¨äº”å…¥å°æ•¸é»
    series_data_month_growth = round(series_data_month_growth)
    # ç›®å‰åªæœ‰ä¸ƒå…«æœˆæœ‰è³‡æ–™
    series_data_month_growth = series_data_month_growth.iloc[[6,7],:]

    # è¨ˆç®—æ¯å€‹æœˆçš„å¹´å¢é•·
    series_data_month_growth['å¹´å¢æˆé•·'] = round( (series_data_month_growth[compare_year2] - series_data_month_growth[compare_year1] )/ series_data_month_growth[compare_year1], 2)
    # è¨ˆç®—ã€Œç¸½å¹´å¢æˆé•·ç‡ã€ï¼šäº†è§£ä»Šå¹´æ¯”è¼ƒå»å¹´ä¾†èªªï¼Œå¢é•·äº†å¤šå°‘
    series_data_month_growth['ç¸½å¹´å¢æˆé•·ç‡'] = (series_data_month_growth[compare_year2].sum() - series_data_month_growth[compare_year1].sum())/ series_data_month_growth[compare_year1].sum()
    series_data_month_growth.to_csv( 'Top' +  str(rank) +'_é«˜PVé »é“_ã€Œ' + series + 'ã€_ã€Œæ½›åœ¨ã€æˆé•·æ¯”è¼ƒåˆ†æè¡¨.csv', encoding='utf-8-sig')
     # ç¹ªè£½ã€Œæ½›åœ¨ã€æˆé•·æ¯”è¼ƒåˆ†æåœ–



    # rank = 1 
    # series = 'ç³»åˆ—3'
    # å†æ¬¡ç¹ªè£½æŠ˜ç·šåœ–ï¼š
    fig = px.line(series_data_month_growth, 
                x="æœˆä»½", 
                y="å¹´å¢æˆé•·", 
                #   color = 'è¨‚å–®æ™‚é–“_å¹´ä»½',
                title ='Top' +  str(rank) +'_é«˜ç²åˆ©ç³»åˆ—_ã€Œ' + series + 'ã€_ã€Œæ½›åœ¨ã€æˆé•·æ¯”è¼ƒåˆ†æåœ–')

    # å¦‚ä½•åŠ å…¥èƒŒæ™¯å…ƒç´ ï¼ˆlayoutï¼‰
    fig.update_layout(
                    title={
                        'text':  'Top' +  str(rank) +'_é«˜ç²åˆ©ç³»åˆ—_ã€Œ' + series + 'ã€_ã€Œæ½›åœ¨ã€æˆé•·æ¯”è¼ƒåˆ†æåœ–',
                        'y':0.9,
                        'x':0.5,
                        'xanchor': 'center',
                        'yanchor': 'top'},
                    xaxis_title='æœˆä»½',
                    yaxis_title='å¹´å¢æˆé•·')
    fig.update_traces(mode="lines+markers")
    plot(fig, filename = 'Top' +  str(rank) +'_é«˜ç²åˆ©ç³»åˆ—_ã€Œ' + series + 'ã€_ã€Œæ½›åœ¨ã€æˆé•·æ¯”è¼ƒåˆ†æåœ–.html')

    # rank = rank +1
    return series_data_month_growth, test

```
